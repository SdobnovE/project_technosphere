```python
mystem = Mystem() 
russian_stopwords = stopwords.words("russian")
```

```python
reg = re.compile(r'\d')
```

```python
path = "./content/content/"
files = os.listdir(path=path) 
text = ""


def get_text(num_doc):
    
    with open(path + str(num_doc) + ".dat", "r") as f:
        text = f.readlines()
        
    bs = BeautifulSoup("".join(text).replace("\n", ""))
    
    h1 = bs.find_all("h1")
    h2 = bs.find_all("h2")
    h3 = bs.find_all("h3")
    a = bs.find_all("a")
    
    res = list(map(lambda x: x.text + " ", a))
    res += list(map(lambda x: x.text + " ", h1))
    res += list(map(lambda x: x.text + " ", h2))
    res += list(map(lambda x: x.text + " ", h3))
    
    
    return "".join(res)

def delete_symbols(text):
    res = ""
    text = text.lower()
    
    for i in text:
        t = i    
        if not(
                (t >= '0' and t <= '9') or 
                (t >= 'a' and t <= 'z') or
                (t >= 'а' and t <= 'я')
              ):
            t = ' '
            
        res += t
        
    return res 

def return_main_k(tokens, k):
    tok, num = np.unique(tokens,return_counts=True)
    if k >= tok.shape[0]:
        return tok
    ind = np.argpartition(num, -k)[-k:]
    return tok[ind]

def lemmatize(number, k):
    text = get_text(number)
    text = delete_symbols(text)
    
    tokens = []
    for i in text.split():
        tokens.append(mystem.lemmatize(i)[0])
    
    tokens_new = []
    
    for word in tokens:
        if word not in russian_stopwords and len(reg.findall(word)) == 0:
            tokens_new.append(word)
    tokens = tokens_new
    tokens = return_main_k(tokens, k)
    
    res = ""
    for i in tokens:
        res += i + " "
    return res

def lemmatize_text(text):
    
    text = delete_symbols(text)
    
    tokens = []
    for i in text.split():
        tokens.append(mystem.lemmatize(i)[0])
    
    tokens_new = []
    
    for word in tokens:
        if word not in russian_stopwords and len(reg.findall(word)) == 0:
            tokens_new.append(word)
    tokens = tokens_new
    
    
    res = ""
    for i in tokens:
        res += i + " "
    return res
```
